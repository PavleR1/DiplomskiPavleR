```python
import pandas as pd
import time
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats,commonplayerinfo
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
nba_players = players.get_players()
```

```python
nba_players_df = pd.DataFrame(data=nba_players)
nba_players_df
```

```python
nba_players_id = [x['id'] for x in nba_players]
```

```python
players_df = pd.DataFrame()
players_df = players_df.append(playercareerstats.PlayerCareerStats(player_id=nba_players_id[0]).get_data_frames()[0])
```

```python
i = 0
for x in range(0,len(nba_players_id)):
    players_df = players_df.append(playercareerstats.PlayerCareerStats(player_id=nba_players_id[x]).get_data_frames()[0])
    if i ==40:
        time.sleep(10)
        i=0
    else:
        i+=1
```

```python
players_df
```

```python
common_players_df = commonplayerinfo.CommonPlayerInfo(player_id=nba_players_id[0]).get_data_frames()[0]
```

```python
i=0
for x in range(0,len(nba_players_id)):
    common_players_df = common_players_df.append(commonplayerinfo.CommonPlayerInfo(player_id=nba_players_id[x]).get_data_frames()[0])
    if i ==20:
        time.sleep(10)
        i=0
    else:
        i+=1
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
commonplayerinfo.CommonPlayerInfo(player_id=2431).get_data_frames()[0]
```

```python
players_df = players_df.merge(nba_players_df,left_on='PLAYER_ID',right_on='id')
```

```python
players_df.head()
```

```python
players_df = players_df.drop(['id','first_name','last_name'],1)
```
